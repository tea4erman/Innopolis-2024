```python
import numpy as np
import pandas as pd
import seaborn as sns
import statsmodels.api as sm
import matplotlib.pyplot as plt
from scipy import stats
```

### Про функциональные формы зависимости

```python
df_bikes = pd.read_csv(r"https://raw.githubusercontent.com/YBI-Foundation/Dataset/refs/heads/main/Bike%20Prices.csv")
```

```python
df_bikes.head()
```

```python
stats.pearsonr(df_bikes['Selling_Price'], df_bikes['Year'])
```

Коэффициент линейной корреляции не высокий

```python
fig, ax = plt.subplots(1,1, figsize = (10,5))
sns.scatterplot(x = 'Year', y = 'Selling_Price', data = df_bikes, ax = ax)
plt.show()
```

По точечной диаграмме видно:

1) Есть выбросы (несколько мотоциклов старше 1995 года) и два мотоцикла дороже 500000 - лучше их убрать

2) Для оставшихся наблюдений зависимость напоминает правую ветвь параболы - скорее всего функциональная форма зависимости должна быть не линейная, а квадратичная

Создадим переменную Age как разницу 2020 и Year

```python
df_bikes['Age'] = 2020 - df_bikes['Year']
```

```python
df_bikes = df_bikes[(df_bikes['Selling_Price']<500000)&(df_bikes['Year']>=2000)]
fig, ax = plt.subplots(1,1, figsize = (10,5))
sns.scatterplot(x = 'Age', y = 'Selling_Price', data = df_bikes, ax = ax)
plt.show()
```

На оставшихся после удаления выбросов наблюдениях видно, что взаимосвязь не линейная

Оценим уравнение линейной регрессии

```python
import statsmodels.formula.api as smf
```

```python
model = smf.ols(formula = 'Selling_Price ~ Age', data = df_bikes)
results = model.fit()
print(results.summary()) # Print the table
```

Уравнение линейной зависимости между возрастом и ценой: $Price_i = 94850 - 6103.97 \cdot Age_i$

Для нового мотоцикла в 2020 году средняя цена составляет 94850

С каждый годом в среднем мотоциклы теряют в цене 6103,97 д.е.

Одним из показателей качества модели регрессии является R-squared $R^2$

Его значение может находиться в интервале от 0 до 1

Его интерпретация - это какой процент изменений в результирующей переменной способны объяснить включенные в модель факторы

В нашем случае, возраст мотоцикла способен объяснить разницу в ценах мотоциклов на 25.3%

Попытаемся в модель включить нелинейность взаимосвязи между ценой и возрастом. Это можно сделать несколькими способами:

1) Оценить модель типа $Price_i = \beta_0 + \beta_1 \cdot Age_i + \beta_2 \cdot Age^2_i$

2) Оценить модель типа $Price_i = \beta_0 + \beta_1 \cdot \frac{1}{Age_i} $

3) Оценить модель типа $Price_i = \beta_0 + \beta_1 \cdot log{Age_i}$

```python
df_bikes['Age2'] = df_bikes['Age']**2
df_bikes['Age_inv'] = 1/(1 + df_bikes['Age'])
df_bikes['log_age'] = np.log(1 + df_bikes['Age'])
```

```python
from statsmodels.iolib.summary2 import summary_col
```

```python
results1 = smf.ols(formula = 'Selling_Price ~ Age + Age2', data = df_bikes).fit()
results2 = smf.ols(formula = 'Selling_Price ~ Age_inv', data = df_bikes).fit()
results3 = smf.ols(formula = 'Selling_Price ~ log_age', data = df_bikes).fit()
```

```python
sum_table = summary_col([results, results1, results2, results3],
                       regressor_order = ['Age', 'Age2', 'Age_inv', 'log_age'],
                       model_names = ['linear', 'quadratic', 'inverse', 'logarithmic'],  float_format = '%.2f')
print(sum_table)
```

На основе показателя R-squared можно сказать, что наилучшим образом зависимость между возрастом мотоцикла и ценой описывает модель:

$Price_i = 138242.31 - 44884.51 \cdot log{Age_i}$

```python
stats.pearsonr(df_bikes['Selling_Price'], df_bikes['log_age'])
```

```python
model4 = smf.ols(formula = 'Selling_Price ~ log_age + KM_Driven', data = df_bikes)
results4 = model4.fit()
print(results4.summary()) # Print the table
```

Добавление переменной "Пробег" не дает повышения качества модели => эта переменная в модели не нужна.
